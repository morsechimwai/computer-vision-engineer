### Prepare Environment
- OpenCV (opencv-python)
- Pillow (pillow)
- Matplotlib (matplotlib)

### Read Image

#### OpenCV
- read image to array, default colors = BGR

```python
import cv2
import matplotlib.pyplot as plt

# read image
img = cv2.imread("images/lenna.png")

# convert BGR → RGB to display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

#### Pillow
- read image using pillow

```python
from PIL import Image

img_pil = Image.open("images/lenna.png")
img_pil.show()
```

### Image Resizing

#### OpenCV

```python
resized = cv2.resize(img_rgb, (200, 200))  # กำหนดขนาดใหม่
plt.imshow(resized)
plt.axis("off")
plt.show()
```

#### Pillow

```python
resized_pil = img_pil.resize((200, 200))
resized_pil.show()
```

### Filtering (Blur/Sharpen/Edge Detection)

#### OpenCV

```python
# Gaussian Blur
blur = cv2.GaussianBlur(img_rgb, (5,5), 0)

plt.subplot(1,2,1)
plt.title("Blur")
plt.imshow(blur)
plt.axis("off")
```

```python
# Edge Detection (Canny)
edges = cv2.Canny(img_rgb, 100, 200)

plt.subplot(1,2,2)
plt.title("Edges")
plt.imshow(edges, cmap="gray")
plt.axis("off")
plt.show()
```

#### Pillow
- using ImageFilter

```python
from PIL import ImageFilter

blur_pil = img_pil.filter(ImageFilter.GaussianBlur(5))
sharpen_pil = img_pil.filter(ImageFilter.SHARPEN)

blur_pil.show()
sharpen_pil.show()
```
